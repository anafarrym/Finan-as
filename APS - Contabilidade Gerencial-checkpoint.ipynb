Empresas escolhidas: Guararapes e Renner

```python
#importando bibliotecas

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

guararapes = pd.read_excel("guararapes.xlsx")
renner = pd.read_excel("rener.xlsx")
```

```python

#cpv
print(guararapes.iloc[193])
#desp com vendas
print((guararapes.iloc[196]))
#desp adm
print((guararapes.iloc[197]))
#print(guararaes.iloc[])
```

### Análise da Renner - LREN3

```python
despesas_totais_lren = renner.iloc[193:195,0:44].sum() 
despesas_totais_lren = pd.DataFrame(despesas_totais_lren) 
despesas_totais_lren.reset_index(drop=True, inplace=True)
despesas_totais_lren = despesas_totais_lren.rename(columns=despesas_totais_lren.iloc[0]).loc[1:].reset_index()
despesas_totais_lren.head()
```

```python
custos_lren = renner.iloc[190,0:44]
custos_lren = pd.DataFrame(custos_lren)
custos_lren.reset_index(drop=True, inplace=True)
custos_lren = custos_lren.rename(columns=custos_lren.iloc[0]).loc[1:].reset_index()
custos_lren.head()
```

```python
receitas_lren = renner.iloc[189,0:44]
receitas_lren = pd.DataFrame(receitas_lren)
receitas_lren.reset_index(drop=True, inplace=True)
receitas_lren = receitas_lren.rename(columns=receitas_lren.iloc[0]).loc[1:].reset_index()
#print(receitas_lren)
receitas_lren.head()
```

```python
#primeira junção
juncao_lren = receitas_lren.merge(custos_lren, how='right', on='index')
base_lren = juncao_lren.merge(despesas_totais_lren, how='right', on='index')
base_lren.head()
```

```python
#rodando a regressão
X_lren = base_lren[['-Custo Produtos Vendidos', '+Despesas com Vendas+Despesas administrativ']]
y_lren = base_lren['+Receita liquida operac']

# Criando e ajustando o modelo de regressão linear
model_lren = LinearRegression()
model_lren.fit(X_lren, y_lren)

# Coeficientes e intercepto
coeficientes_lren = model_lren.coef_
intercepto_lren = model_lren.intercept_

# Imprimindo os coeficientes e intercepto
print("Coeficientes:", coeficientes_lren)
print("Intercepto:", intercepto_lren)
```

```python
#cálculo GAO
GAO_lren = 2.53791177 + -0.4308029
print(GAO_lren)
```

```python
# Fazer previsões com base nos dados de entrada
y_pred_lren = model_lren.predict(X_lren)
y_pred_lren = pd.DataFrame(y_pred_lren)
y_pred_lren.head()
```

```python
#plotando comparações
# criando a figura e os eixos
fig, ax = plt.subplots()
fig.set_facecolor('white')
ax.set_facecolor('white')

# criando o segundo eixo Y
ax2 = ax.twinx()

# plotando os dados no primeiro eixo Y
ax.plot(receitas_lren, label='Receitas Realizadas', color='red')
ax.set_xlabel('valores')
ax.set_ylabel('Receitas Realizadas')

# plotando dados no segundo eixo Y
ax2.plot(receitas_lren.index, y_pred_lren, label='Receitas previstas', color='orange')
ax2.set_ylabel('Receitas previstas')

# adicionando a legenda
ax.legend(loc='upper left')
ax2.legend(loc='upper right')

# adicionando o título
plt.title('Comparação das receitas LREN3')

# exibindo o gráfico
plt.show()
```
